```python
# %OEE by Orden
%%sql
SELECT Orden, "Disponibilidad (%)", "Rendimiento (%)", "Calidad (%)", AVG(ROUND(("Disponibilidad (%)" * "Rendimiento (%)" * "Calidad (%)"),2)) AS "OEE (%)"
FROM Data_Modelo
GROUP BY Orden
ORDER BY "OEE (%)" DESC
LIMIT (100)
```

```python
# %OEE by Material
%%sql
SELECT Material, ROUND(AVG("Disponibilidad (%)"),2), ROUND(AVG("Rendimiento (%)"),2), ROUND(AVG("Calidad (%)"),2), ROUND(AVG(("Disponibilidad (%)" * "Rendimiento (%)" * "Calidad (%)")),2) AS "OEE (%)"
FROM Data_Modelo
GROUP BY Material
ORDER BY "OEE (%)" DESC
```

```python
# %OEE by Linea
%%sql
SELECT Linea, ROUND(AVG("Disponibilidad (%)"),2), ROUND(AVG("Rendimiento (%)"),2), ROUND(AVG("Calidad (%)"),2), ROUND(AVG(("Disponibilidad (%)" * "Rendimiento (%)" * "Calidad (%)")),2) AS "OEE (%)"
FROM Data_Modelo
GROUP BY Linea
ORDER BY "OEE (%)" DESC
```

```python
# %OEE mayor o igual al promedio, desglosado por Material
%%sql
SELECT *
FROM(SELECT Material, ROUND(AVG("Disponibilidad (%)"),2), ROUND(AVG("Rendimiento (%)"),2), ROUND(AVG("Calidad (%)"),2), ROUND(AVG(("Disponibilidad (%)" * "Rendimiento (%)" * "Calidad (%)")),2) AS "OEE (%)"
      FROM Data_Modelo
      GROUP BY Material
      ORDER BY "OEE (%)" DESC)
WHERE "OEE (%)" >= (SELECT AVG("OEE (%)")
                    FROM(SELECT Material, ROUND(AVG("Disponibilidad (%)"),2), ROUND(AVG("Rendimiento (%)"),2), ROUND(AVG("Calidad (%)"),2), ROUND(AVG(("Disponibilidad (%)" * "Rendimiento (%)" * "Calidad (%)")),2) AS "OEE (%)"
                          FROM Data_Modelo
                          GROUP BY Material
                          ORDER BY "OEE (%)" DESC))
```
